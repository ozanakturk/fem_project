```python
jk = hoo
for i in range(0, 1):
    jk = np.concatenate((jk, hoo), axis=1)
N = jk.shape[1]
jj = np.fft.fft(jk)

# ll = -np.sort(-abs(jj[:,:300]), axis=1)
ll = -np.sort(-abs(jj[:,np.int(np.round(jj.shape[1]/2)):]), axis=1)
```

```python
plt.plot((ll[:,0]))
```

```python
ya = np.linspace(0, 6, ll.shape[0])
```

```python
%matplotlib inline
import matplotlib
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(12, 8))
matplotlib.rcParams.update({'font.size': 22})
plt.rc('font', family='serif')
plt.rc('text', usetex=True)
ax = fig.gca()
# plt.plot(np.abs(jj[:, 0]))
plt.plot(ya, ll[:, 0] * 2/N / 1e6)
plt.plot(ya, ll[:, 1] * 2/N / 1e6)
plt.plot(ya, ll[:, 2] * 2/N / 1e6)
plt.plot(ya, ll[:, 3] * 2/N / 1e6)
# plt.plot(ll[:,2] * 2/N)
# plt.plot(ll[:,3] * 2/N)
plt.legend(('First harmonic', 'Second hamonic', 'Third harmonic', 'Fourth harmonic'),
           shadow=True, loc=(0.65, 0.72), handlelength=1.5, fontsize=20)
plt.grid(True)
plt.ylim([0, 2])
plt.xlim([0, 6])
plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1, 1.2, 1.4, 1.6, 1.8, 2])
# plt.plot(y_line, u_total_periods)
plt.xlabel(r'$x$ (cm)')
plt.ylabel('Pressure (MPa)')
fig.savefig('plane_circular_piston_fenics_p2_nPerLam6.pdf')
```
